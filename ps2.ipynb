```python
####################
#### Problem 3 plots 3a,c,d,e   #####
###################

#X0 = 1 // it is 1 to match the plot

def x0(t): #x0 is the undamped
    return np.cos(2*t)

def x1(t):  #equation c: underdamped
    d=np.sqrt(3)
    return np.exp(-t)*(np.cos(d*t)+(np.sin(d*t)/d))

def x2(t): #equation d: overdamped
    u=(np.sqrt(5)+3)/(2*np.sqrt(5))
    q=(np.sqrt(5)-3)/(2*np.sqrt(5))
    r=np.sqrt(5)
    return np.exp(-3*t)*(u*np.exp(r*t)+q*np.exp(-r*t))

def x3(t): #equation e: critically damped 
    return np.exp(-2*t)+2*t*np.exp(-2*t)
    

t = np.linspace(0,10,100)
tnew = np.linspace(np.pi, np.pi,100)
y = np.linspace(-10,10,100)

plt.plot(t,x0(t), label="undamped")
plt.plot(t,x1(t), label="underdamped")
plt.plot(t,x2(t), label="overdamped")
plt.plot(t,x3(t), label="critcally damped")
plt.plot(tnew,y, "--", label="1 period") #1 period = pi

plt.xlabel("time")
plt.ylabel("position")
plt.legend()

plt.ylim(-1,1)
```
