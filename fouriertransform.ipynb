```python
#Fourier Transform and Fourier Series in Python
#Code Explanation
#Wave Generation:

#The code generates three different types of waves:
#Sine Wave: A simple sine wave at 5 Hz.
#Square Wave: A square wave generated at 5 Hz.
#Sawtooth Wave: A sawtooth wave generated at 5 Hz.
#Fourier Transform:

#For each wave, we compute the Fourier Transform using np.fft.fft, which transforms the time-domain signal into its frequency-domain representation.
#The Fourier Transform outputs the magnitude of the frequency components, which we plot to visualize the frequency spectrum.
#Fourier Series Approximation of a Square Wave:

#The code also demonstrates how a square wave can be approximated using a Fourier series.
#This progression is shown by adding more terms to the Fourier series, improving the approximation with each additional harmonic.
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# Define the time domain (0 to 2π) and sampling rate
sampling_rate = 1000  # samples per second
T = 2 * np.pi  # period (0 to 2π)
t = np.linspace(0, T, sampling_rate)  # time array

# Function to approximate a square wave using Fourier series
def square_wave_fourier_series(t, n_terms):
    approximation = np.zeros_like(t)
    for n in range(1, n_terms * 2, 2):  # odd harmonics only
        approximation += (4 / np.pi) * np.sin(n * t) / n
    return approximation

# Define waveforms: sine, square, and sawtooth
freq = 5  # Frequency in Hz
sine_wave = np.sin(2 * np.pi * freq * t)  # 5 Hz sine wave
square_wave = signal.square(2 * np.pi * freq * t)  # 5 Hz square wave
sawtooth_wave = signal.sawtooth(2 * np.pi * freq * t)  # 5 Hz sawtooth wave

# Plot time domain signals
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, sine_wave)
plt.title('Sine Wave (5 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, square_wave)
plt.title('Square Wave (5 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, sawtooth_wave)
plt.title('Sawtooth Wave (5 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')

plt.tight_layout()
plt.show()

# Compute Fourier Transforms
sine_wave_fft = np.fft.fft(sine_wave)
square_wave_fft = np.fft.fft(square_wave)
sawtooth_wave_fft = np.fft.fft(sawtooth_wave)

# Get corresponding frequencies
frequencies = np.fft.fftfreq(len(t), t[1] - t[0])

# Plot Fourier Transforms (magnitude spectrum)
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frequencies[:len(frequencies)//2], np.abs(sine_wave_fft[:len(t)//2]))
plt.title('Fourier Transform of Sine Wave (5 Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.subplot(3, 1, 2)
plt.plot(frequencies[:len(frequencies)//2], np.abs(square_wave_fft[:len(t)//2]))
plt.title('Fourier Transform of Square Wave (5 Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.subplot(3, 1, 3)
plt.plot(frequencies[:len(frequencies)//2], np.abs(sawtooth_wave_fft[:len(t)//2]))
plt.title('Fourier Transform of Sawtooth Wave (5 Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')

plt.tight_layout()
plt.show()

# Visualize progression of Fourier series approximation of a square wave
n_values = [1, 3, 5, 11, 21]  # different numbers of terms to show progression
plt.figure(figsize=(10, 10))

for i, n in enumerate(n_values, 1):
    plt.subplot(len(n_values), 1, i)
    plt.plot(t, square_wave_fourier_series(t, n))
    plt.title(f'Fourier Series Approximation of Square Wave with {n} Terms')
    plt.ylim(-1.5, 1.5)
    plt.grid()

plt.tight_layout()
plt.show()
```
