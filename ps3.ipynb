```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import math as mp
```

```python
## part (a) The first cell ---  Define a sinusoid:
omega = 0.5
phi = 0
A = 3.0
t = np.arange(0,10,0.1)
x = A*np.sin(2*np.pi*omega*t + phi)

plt.plot(t,x)
plt.xlabel('Time [s]')
plt.ylabel('height [m]')
plt.title('Input data')
plt.show()
```

```python
# part (b) The second cell -- decrease the sample rate to the values given and see what happens
## part (a) The first cell ---  Define a sinusoid:
omega = 0.5
phi = 0
A = 3.0

# Sampling rate: 0.1s
t = np.arange(0,10,0.1)
x = A*np.sin(2*np.pi*omega*t + phi)
plt.plot(t,x)

# Sampling rate: 0.2s
t = np.arange(0,10,0.2)
x = A*np.sin(2*np.pi*omega*t + phi)
plt.plot(t,x)

# Sampling rate: 0.5s
t = np.arange(0,10,0.5)
x = A*np.sin(2*np.pi*omega*t + phi)
plt.plot(t,x)

# Sampling rate: 0.6s
t = np.arange(0,10,0.6)
x = A*np.sin(2*np.pi*omega*t + phi)
plt.plot(t,x)

# Sampling rate: 1.0s
t = np.arange(0,10,1.0)
x = A*np.sin(2*np.pi*omega*t + phi)
plt.plot(t,x)

# Sampling rate: 1.5s
t = np.arange(0,10,1.5)
x = A*np.sin(2*np.pi*omega*t + phi)
plt.plot(t,x)

plt.legend(['S_R=0.1s', 'S_R=0.2s', 'S_R=0.5s','S_R=0.6s','S_R=1.0s','S_R=1.5s'])
plt.xlabel('Time [s]')
plt.ylabel('height [m]')
plt.title('Input data')
plt.show()
```

```python
omega = 0.5
phi = 0
A = 3.0
t = np.arange(0,10,0.1)
x = A*np.sin(2*np.pi*omega*t + phi)

## part (c) Take its fourier transform
N=len(x); # total number of sample
Fs=1/0.1; # sampling frequency 
x_ft = np.abs(np.fft.fft(x,N)) # fft of the singal

# create x axis for frequency
Fr=(Fs/N)*np.arange(N)
plt.plot(Fr,x_ft)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of input data')
plt.show();
```

```python
## part (d) Take its fourier transform

N=len(x); # total number of sample
Fs=1/0.1;  # sampling frequency 
x_ft = np.abs(np.fft.fft(x)) # fft of the singal
x_ft_half=x_ft[0:(51)]
# create frequency array and cut in half
#Fr=(Fs/(2*N)*np.arange(N/2);
Fr_half=np.linspace(0,Fs/2,(mp.floor(N/2)+1)) ## Frequency array in half 
plt.plot(Fr_half,x_ft_half,'b-o')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of input data')
plt.show()
```

```python
## part (d) Take its fourier transform

N=len(x); # total number of sample
Fs=1/0.1; # samplinf frequency 
x_ft = np.abs(np.fft.fft(x,N))
x_ft_mag=(2/N)*x_ft[0:(51)]; # fft of the singal in half and normalized
# create frequency array and cut in half
#Fr=(Fs/(2*N)*np.arange(N/2);
Fr=np.linspace(0,Fs/2,(mp.floor(N/2)+1))    ## Frequency array
plt.plot(Fr,x_ft_mag,'b-o')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of input data')
plt.show()
```

```python
## part (f) 
data = np.loadtxt('oceandata.txt')
data.shape
times = data[:,0]
heights = data[:,1]

plt.plot(times,heights)
#plt.xlim(0,10)
plt.show()

dt=times[2]-times[1]

N=len(heights); # total number of sample
Fs=1/dt; # samplinf frequency 
x_ft = np.abs(np.fft.fft(heights))
x_ft_mag=(2/N)*x_ft[0:(mp.floor(N/2)+1)]; # fft of the singal in half and normzied
# create frequency array and cut in half
#Fr=(Fs/(2*N)*np.arange(N/2);
Fr=np.linspace(0,Fs/2,(mp.floor(N/2)+1))    ## Frequency array
plt.plot(Fr,x_ft_mag,'b-o')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of input data with 0th bin')
plt.show()

# excluding the 0th bin 
x_ft_mag[0]=0;
plt.plot(Fr,x_ft_mag,'b-o')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of input data excluding 0th bin')
plt.show()
```
